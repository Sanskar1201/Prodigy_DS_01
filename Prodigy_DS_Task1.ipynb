Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

#Data Loading and Inspection:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/HR-Employee-Attrition.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

#Visualizations:

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Visualize the distribution of Gender using a bar chart
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Distribution of Gender in Employee Attrition', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Visualize the distribution of Age using a histogram
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Age in Employee Attrition', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
#Visualize the distribution of Department using a bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=df, palette='Set2')
plt.title('Distribution of Department in Employee Attrition', fontsize=16)
plt.xlabel('Department', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Visualize the distribution of Monthly Income using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyIncome'], bins=20, kde=True, color='salmon', edgecolor='black')
plt.title('Distribution of MonthlyIncome in Employee Attrition', fontsize=16)
plt.xlabel('Monthly Income', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

#Conclusion:

The visualizations provide a comprehensive overview of key variables in the employee dataset, including gender, age, department distribution, and monthly income. These visualizations serve as valuable tools for exploring the dataset and gaining insights into the characteristics of the employees, which can be crucial for further analysis and decision-making.
